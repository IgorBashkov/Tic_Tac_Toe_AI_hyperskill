```python
from random import choice
from itertools import cycle

def win(char):
    # function defines winner
    # it checks all rows, columns and diagonals for same symbols
    global data_cell
    matr = data_cell
    char *= 3
    for i in matr:
        if ''.join(i) == char:
            return True
    for i in range(3):
        if ''.join([x[i] for x in matr]) == char:
            return True
    if ''.join([matr[i][i] for i in range(3)]) == char:
        return True
    if ''.join([matr[i][2 - i] for i in range(3)]) == char:
        return True
    return False


def print_cell(matr):
    # function prints curret game field with border
    out_matr = ['-' * 9] * 5
    for i in range(3):
        out_matr[i+1] = '| ' + ' '.join(matr[i]) + ' |'
    for i in range(len(out_matr)):
        print(out_matr[i])

        
def change_cell(matr, x, y, ch='X'):
    # function check cell for void and change it
    x -= 1
    y = -y
    if matr[y][x] == ' ':
        matr[y][x] = ch
        flag = False
    else:
        print('This cell is occupied! Choose another one!')
        flag = True
    return matr, flag


def get_coord(row_list):
    # function checks input for errors
    if len(row_list) != 2:
        print('You should enter numbers!')
        return row_list, True
    try:
        row_list = [int(i) for i in row_list]
    except ValueError:
        print('You should enter numbers!')
        return row_list, True
    if (1 > row_list[0] or row_list[0] > 3) or (1 > row_list[1] or row_list[1] > 3):
        print('Coordinates should be from 1 to 3!')
        return row_list, True
    return row_list, False 


def easy_rand():
    global data_cell
    while True:
        X = choice((1, 2, 3))
        Y = choice((1, 2, 3))    
        x = X - 1
        y = -Y
        if data_cell[y][x] == ' ':
            print('Making move level "easy"')
            return (get_coord([X, Y]))
        
def check_seq(ch):
    global data_cell
    matr = data_cell
    
    for n, i in enumerate(matr):
        if i.count(ch) == 2 and i.count(' ') == 1:
            for k, j in enumerate(i):
                if j == ' ':
                    
                    return True, [k, n]
                
    for i in range(3):
        tmp_lst = [x[i] for x in matr]
        if  tmp_lst.count(ch) == 2 and tmp_lst.count(' ') == 1:
            for n, j in enumerate([x[i] for x in matr]):
                if j == ' ':
                    
                    return True, [i, n]

    tmp_lst = [matr[i][i] for i in range(3)]
    if tmp_lst.count(ch) == 2 and tmp_lst.count(' ') == 1:
        for n, j in enumerate([matr[i][i] for i in range(3)]):
            if j == ' ':
                
                return True, [n, n]
    
    tmp_lst = [matr[i][2 - i] for i in range(3)]
    if tmp_lst.count(ch) == 2 and tmp_lst.count(' ') == 1:
        for n, j in enumerate([matr[i][2 - i] for i in range(3)]):
            if j == ' ':
                
                return True, [2-n, n]
            
    return False, [0, 0]
        
def medium_move(ch):
    ch_1 = 'XO'.replace(ch, '')
    print(ch_1)
    fl, x_y = check_seq(ch)
    
    if fl:
        print('Making move level "medium"')
        return [x_y[0] + 1, 3 - x_y[1]], False
    
    fl, x_y  = check_seq(ch_1)
    
    if fl:
        print('Making move level "medium"')
        return [x_y[0] + 1, 3 - x_y[1]], False
        
    global data_cell
    
    while True:
        X = choice((1, 2, 3))
        Y = choice((1, 2, 3))    
        x = X - 1
        y = -Y
        if data_cell[y][x] == ' ':
            print('Making move level "medium"')
            return (get_coord([X, Y]))

    

def check_start():
    global lst
    lst = input('Input command:').split()
    if len(lst) != 1 and len(lst) != 3:
        
        return False
    set_1 = {'start', 'exit'}
    set_2 = {'user', 'easy', 'medium'}
    if lst[0] in set_1:
        if lst[0] == 'start' and len(lst) == 3:
            if set(lst[1:3]).issubset(set_2):
                return True
        elif lst[0] == 'exit' and len(lst) == 1:
            return True                
        
    return False


def wait_input():
    while not check_start():
        print('Bad parameters!')

def player_step(player):
    global seq
    if player == 'user':
        return get_coord(input('Enter the coordinates:').split())
    elif player == 'easy':
        return easy_rand()
    elif player == 'medium':
        return medium_move(seq)

def get_name(name_list):
    for i in cycle(name_list):
        yield i
    
symb = 'XO'

while True:
    wait_input()
    if lst[0] == 'exit':
        break
    player_seq = get_name(lst[1:])
    symb_seq = get_name(list(symb))
    data = ' ' * 9
    data_cell = [[x for x in data[i * 3:(i+1) * 3]] for i in range(3)]
    print_cell(data_cell)
    
    for step in range(9):
        seq = next(symb_seq)
        coord_true = True
        x_y = []
        next_player = next(player_seq)
        while coord_true:
            
            x_y, coord_true = player_step(next_player)

            if not coord_true:
                data_cell, coord_true = change_cell(data_cell, x_y[0], x_y[1], seq)
                
        print_cell(data_cell)

        if step > 3:
            if win('X'):
                print('X wins')
                break
            elif win('O'):
                print('O wins')
                break
        if step == 8:
            print("Draw")
            break
```
